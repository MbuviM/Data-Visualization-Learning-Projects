```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
# plt.style is used to choose appropriate aesthetic styles for the graphs.
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), ('-'))
plt.plot(x, np.cos(x), ('--'))
```

```python
"""Saving the figure as a png file"""
fig.savefig('my_figure.png')
```

## Two Interfaces For the Price of One

MATLAB-style Interface

```python
"""Creation of subplots"""
# Create a plot figure
fig2 = plt.figure()

# Create the first of the two panels
plt.subplot(2, 1, 1) # Rows, columns, panel number
plt.plot(x, np.sin(x))

# Create the second of the two panels
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
fig2.savefig('subplots.png')
```

Object-Oriented Interface

```python
"""Used for more complicated situations and when you want to have more control over the figures"""
# Create a grid of plots
# ax will be an array of two Axes objects
fig3, ax = plt.subplots(3)

# Call plot method on the appropriate objects
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
```

```python
fig3.savefig('subplots_2.png')
```

# Simple Line Plots

```python
plt.style.use('seaborn-whitegrid')
```

```python
# Creating of a figure and its axes
fig4 = plt.figure()
ax = plt.axes()
```

```python
y = np.linspace(0, 10, 1000)
ax.plot(y, np.sin(y))
```

```python
# Adjusting the color and graph style
plt.plot(y, np.sin(y), color='red', linestyle='-') #solid
plt.plot(y, np.cos(y), color='blue', linestyle='--') #dashed
```

```python
plt.plot(y, np.sinh(y), color='brown', linestyle='-.') #dashdot
plt.plot(y, np.cosh(y), color='green', linestyle=':') #dotted
plt.plot(y, np.tanh(y), color='orange')
```

```python
# Adjusting of Axes Limits using plt.xlim() and plt.ylim()
plt.plot(y, np.sin(y), color='brown', linestyle=':')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# Ajusting the plot in reverse order
plt.plot(y, np.sin(y), color='green', linestyle='-.')
plt.xlim(11, -1)
plt.ylim(1.5, -1.5)
```

```python
# Using plt.axis() to set x and y limits using a single call
plt.plot(y, np.sin(y))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
# Using the plt.axis() method to create a tight layout
plt.plot(y, np.sin(y))
plt.axis('tight')
```

```python
# Using the axis method to create an equal layout
plt.plot(y, np.sin(y))
plt.axis('equal')
```

# Labeling Plots

```python
# Using plt.title(), plt.xlabel(), plt.ylabel() to label the graph
plt.plot(y, np.cos(y), color='yellow')
plt.title('Cosine Curve')
plt.xlabel('y')
plt.ylabel('cos(y)')
```

```python
# Using plt.legend() method to label multiple graphs in an axes
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':r', label='cos(x)')
plt.axis('equal')
plt.legend()
```

```python
# plt.scatter() is the method used to scatter the plot.
# plt.show() is the method used to show the plot.
price = [2.50, 1.23, 4.02, 3.25, 5.00, 4.40]
sales_per_day = [34, 62, 49, 22, 13, 19]
plt.title('Total Sales Per Day')
plt.scatter(price, sales_per_day)
plt.show() # Mostly used in Python Scripts
```

Customizing Markers in Scatter Plots

```python
# Customizing the size of the markers
price = np.asarray([2.50, 1.23, 4.02, 3.25, 5.00, 4.40])
sales_per_day = np.asarray([34, 62, 49, 22, 13, 19])
profit_margin = np.asarray([20, 35, 40, 20, 27.5, 15])
plt.scatter(x=price, y=sales_per_day, s=profit_margin * 10)
```

```python
# Customizing the colour of the marker
low = (0, 1, 0) # Green
medium = (1, 1, 0) # Yellow
high = (1, 0, 0) # Red

sugar_content = [low, high, medium, medium, high, low]
plt.scatter(x=price, y=sales_per_day, s=profit_margin * 10, c=sugar_content)
```

```python
# Customizing the shape of the Markers
low = (0, 1, 0) # Green
medium = (1, 1, 0) # Yellow
high = (1, 0, 0) # Red

sugar_content = [low, high, medium, medium, high, low]
plt.scatter(x=price, y=sales_per_day, s=profit_margin * 10, c=sugar_content)

price_cereal = np.asarray([1.50, 2.50, 1.15, 1.95])
sales_per_day_cereal = np.asarray([67, 34, 36, 12])
profit_margin_cereal = np.asarray([20, 42.5, 33.3, 18])
sugar_content_cereal = [low, high, medium, low]
plt.scatter(x=price_cereal, y=sales_per_day_cereal, s=profit_margin_cereal * 10, c=sugar_content_cereal, marker='d', alpha=0.5)
plt.title('Sales vs Prices for Orange Drinks and Cereal Bars')
plt.legend(['Orange Drinks', 'Cereal Bars'])
plt.xlabel('Price(Unit Currency)')
plt.ylabel('Average Weekly Sales')
plt.text(3.2, 55, 'Size of Marker=Profit Margin\n' 'Colour of Marker=Sugar Content')
```

Customize the Colourmap and Style

```python
sugar_content_orange = [15, 35, 22, 27, 38, 14]
sugar_content_cereal = [21, 49, 29, 24]

plt.scatter(x=price, y=sales_per_day, s=profit_margin * 10, c=sugar_content_orange, cmap='jet', alpha=0.5) # Orange Juice
plt.scatter(x=price_cereal, y=sales_per_day_cereal, s=profit_margin_cereal * 10, c=sugar_content_cereal, cmap='jet', alpha=0.5, marker=',') # Cereal Bars
plt.title("Sales vs Prices for Orange Drinks and Cereal Bars")
plt.legend(["Orange Drinks", "Cereal Bars"])
plt.xlabel("Price (Currency Unit)")
plt.ylabel("Average weekly sales")
plt.text(
    2.7,
    55,
    "Size of marker = profit margin\n" "Color of marker = sugar content",
)
plt.colorbar()
```
